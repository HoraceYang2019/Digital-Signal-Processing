## <font color='blue'>9-1 Fourier Series </font>
Basic concept: Given an any period function, it can be represented as a series of different frequencies and amplitudes of sin and cos functions. <br>
    The series is named Fourier Series.

https://www.khanacademy.org/science/electrical-engineering/ee-signals/ee-fourier-series/v/ee-fourier-series-intro

### Fourier Series
If a function $x(t)$ is defined between intervals $ –L \leq t \leq L$, then the Fourier Series can be defined as:

\begin{equation}
x(t)=\frac{1}{2}a_0 + \sum_{n=1}^{\infty} [a_n \cos \frac{n \pi t}{L} +  b_n \sin \frac{n \pi t}{L}]. \tag{4}      
\end{equation}  

where

\begin{equation}
a_0=\frac{1}{L} \int_{-L}^{L}x(t)dt, \\
a_n=\frac{1}{L} \int_{-L}^{L}x(t)\cos \frac{n \pi t}{L}dt, \\
b_n=\frac{1}{L} \int_{-L}^{L}x(t)\sin \frac{n \pi t}{L}dt. \tag{5} 
\end{equation} 

In this series, $\frac{1}{2}a_0$ is a constant, and $a_n$ and $b_n$ can be considered as the amplidudes of the nth sinsouid with the corresponding frequency $n \omega$.  <br> 
Therefore, we can decompose the componments of the series into DC componment and AC componment. <br>
That is, $x(t) = DC + AC $ <br>
where <br>
$ DC =\frac{1}{2}a_0 $ <br>
$ AC =\sum_{n=1}^{\infty} [a_n \cos \frac{n \pi t}{L} +  b_n \sin \frac{n \pi t}{L}] $
 

#### Expample: 
If a function is defined as 
\begin{equation}
  x(t) =
    \begin{cases}
      -1, & \text {if}\; -1 < t < 0\\
      1, & \text {if}\; 0 \leq t < 1
    \end{cases}       
\end{equation}


Assume $x(t)$ is a periodic function, it meets $x(t) = x(t + 2)$. 
Hence, it will form a idal square wave if extending left and right to infinite <br>
Period $T = 2 $(or $L = 1$). First, the process for finding terms of Fourier Series is as follows.
\begin{equation}
a_0=\frac{1}{L} \int_{-L}^{L}x(t)dt = \int_{-1}^{1}x(t)dt = \int_{-1}^{0}(-1)dt+\int_{0}^{1}(1)dt =0\\
\end{equation} 

\begin{equation}
\begin{split}
a_n &= \frac{1}{L} \int_{-L}^{L}x(t)\cos \frac{n \pi t}{L}dt\\
    &= \int_{-1}^{1}x(t)\cos {n \pi t}dt \\
    &= \int_{-1}^{0}(-1)\cos {n \pi t}dt + \int_{0}^{1}(1)\cos {n \pi t}dt \\
    &= [-\frac{1}{n \pi} \sin(n \pi t) ]_{-1}^{0} + [\frac{1}{n \pi} \sin(n \pi t) ]_{0}^{1}=0
\end{split}
\end{equation} 

\begin{equation}
\begin{split}
b_n &= \frac{1}{L} \int_{-L}^{L}x(t)\sin \frac{n \pi t}{L}dt\\
    &= \int_{-1}^{1}x(t)\sin {n \pi t}dt \\
    &= \int_{-1}^{0}(-1)\sin {n \pi t}dt + \int_{0}^{1}(1)\sin {n \pi t}dt \\
    &= [\frac{1}{n \pi} \cos(n \pi t) ]_{-1}^{0} + [-\frac{1}{n \pi} \cos(n \pi t) ]_{0}^{1} \\
    &= \frac{2}{n\pi}-\frac{2}{n\pi}\cos(n\pi) \\
    &= \frac{2}{n\pi}[1-(-1)^n]
\end{split}
\end{equation} 

Therefore, the $x(t)$ in Fourier Series will be
\begin{equation}
\begin{split}
x(t) &= \frac{1}{2}a_0 + \sum_{n=1}^{\infty} [a_n \cos \frac{n \pi t}{L} +  b_n \sin \frac{n \pi t}{L}] \\
     &= \sum_{n=1}^{\infty} \frac{2}{n\pi}[1-(-1)^n] \sin (n \pi t) \\
     &= \frac{4}{\pi}\sin (\pi t) +\frac{4}{3\pi}\sin (3\pi t) +...
\end{split}
\end{equation}

The Fourier Transform of a function $x(t)$ can be defined as 
\begin{equation}
X(\omega)=F\{ x(t)\}=\int_{-\infty}^{\infty} x(t)e^{-j \omega t} dt. \tag{6} 
\end{equation} 

The Inverse Fourier Transform of a function $X(\omega)$ can be defined as 
\begin{equation}
x(t)=F^{-1}\{ X(\omega)\}=\frac{1}{2\pi}\int_{-\infty}^{\infty} X(\omega)e^{j \omega t} d\omega. \tag{7} 
\end{equation} 

where $\int_{-\infty}^{\infty} |x(t)|dt$ is convergence with $j=\sqrt{-1}$

#### Example. 2
Find the Fourier Transform of function given as $x(t)=Acos(\omega_0 t+\phi)$, <br>
where A, $\omega_0$, and $\phi$ are amplitude, angular frequency, and phase shift, respectively.

\begin{equation}
\begin{split}
X(\omega) &= F\{ x(t) \} =\int_{-\infty}^{\infty} x(t) e^{-j \omega t} dt \\
          &= \int_{-\infty}^{\infty} A \cos (\omega_0 t) e^{-j \omega t} dt \\
          &= A \int_{-\infty}^{\infty} (\frac { e^{j \omega_0 t} + e^{-j \omega_0 t}}{2})e^{-j \omega t} dt \\
          &= \frac{A}{2}[\int_{-\infty}^{\infty} e^{j \omega_0 t}e^{-j \omega t}dt + \int_{-\infty}^{\infty} e^{-j \omega_0 t}e^{-j \omega t}dt]\\
          &= \frac{A}{2}[F\{ e^{j \omega_0 t} \} + F \{ e^{-j \omega_0 t} \}]\\
          &= A \pi [\delta(\omega-\omega_0)+ \delta(\omega+\omega_0)]
\end{split}
\end{equation} 
where, $F\{e^{j \omega_0 t }\} = 2 \pi \delta (\omega-\omega_0)$

#### Example. 3
If a function is defined as
\begin{equation}
  x(t) =
    \begin{cases}
      A, & \text {if}\; -T/2 < t < T/2\\
      0, & \text {otherwise}\; 
    \end{cases}       
\end{equation}
is called ideal pulse function, find the Fourier transform

\begin{equation}
\begin{split}
X(\omega) &= \int_{-\infty}^{\infty} x(t) e^{-j \omega t} dt = \int_{-T/2}^{T/2} A e^{-j \omega t} dt \\
          &= A \int_{-T/2}^{T/2} e^{-j \omega t} dt \\
          &= A [-\frac{1}{j\omega} e^{-\frac{j\omega T}{2}} + \frac{1}{j\omega} e^{\frac{j\omega T}{2}}] \\
          &= \frac{A}{j\omega} [e^{\frac{j\omega T}{2}} - e^{-\frac{j\omega T}{2}}]\\
          &= \frac{A}{j\omega} [\cos (\frac{\omega T}{2})+j\sin (\frac{\omega T}{2})-\cos (\frac{\omega T}{2})+j\sin (\frac{\omega T}{2})] \\
          &= \frac{A}{\omega}[2\sin (\frac{\omega T}{2})] = AT[\frac{\sin(\frac{\omega T}{2})}{\frac{\omega T}{2}}]\\
          &= AT \cdot sinc(\frac{\omega T}{2 \pi})
\end{split}
\end{equation}

### Fourier Transform Shifting Theorems (傅立葉轉換的平移定理)：
#### (1) First Shifting Theorem (第一平移定理)

If $f$ is a time function, and $F$ is its Fourier Transform. Then,
\begin{equation}
F\{ f(t-t_0)\}= F(\omega) e^{-j \omega t_0}. \tag{8} 
\end{equation}

Where $t_0$ is the shifting time and $t_0 > 0$

\begin{equation}
F\{ f(t-t_0)\}= \int_{-\infty}^{\infty} f(t-t_0) e^{-j \omega t} dt. \tag{9} 
\end{equation}
let $\tau = t - t_0 $, then $t=\tau + t_0$, and $d\tau=dt$ <br>
we have 

\begin{equation}
\begin{split}
F\{ f(t-t_0)\} &= \int_{-\infty}^{\infty} f(\tau) e^{-j \omega (\tau+t_0)} d\tau\\
               &= \int_{-\infty}^{\infty} f(\tau) e^{-j\omega \tau} \cdot e^{-j\omega t_0} d\tau\\
               &= F(\omega) \cdot e^{-j\omega t_0}
\end{split}. \tag{10} 
\end{equation} 

#### (2)The Second Shifting Theorem (第二平移定理)

If $f$ is a time function, and $F$ os its Fourier Transform. Then,
\begin{equation}
F\{ f(t) \cdot e^{j\omega_0 t}\}= F(\omega - \omega_0). \tag{11} 
\end{equation}

where $\omega_0$ is the shifting angular Frequency (平移角頻率) and $ \omega_0 > 0$

\begin{equation}
F\{ f(t) \cdot e^{j\omega_0 t}\}= \int_{-\infty}^{\infty}f(t)\cdot e^{j \omega_0 t} \cdot e^{-j \omega t} dt
= \int_{-\infty}^{\infty}f(t)\cdot e^{-j (\omega - \omega_0)t} dt = F(\omega - \omega_0). \tag{12} 
\end{equation}

#### (3)捲積定理

若$f$與$g$為兩個時間的函數，且$F$為傅立葉轉換，則：

\begin{equation}
F\{ f*g \}= F\{f\}*F\{g\}. \tag{13} 
\end{equation}

其中，星號*為卷積運算。


#### (4)高斯函數的傅立葉轉換

高斯函數的傅立葉轉換，形成另一個高斯函數：
\begin{equation}
F\{\ e^{-\frac{t^2}{2 \sigma^2}}\}= \sqrt{2\pi \sigma^2} \cdot e^{-\frac{t^2}{2 \sigma^2}}. \tag{14} 
\end{equation} 

其中，$\sigma$為標準差。

## <font color='blue'>9-3 離散時間傅立葉轉換 </font>
給定離散的序列$x[n]$，則離散時間傅立葉轉換(Discrete-Time Fourier Transform, DTFT)可以定義為：
 
\begin{equation}
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n]e^{-j \omega n}. \tag{16} 
\end{equation}

反離散時間傅立葉轉換(Inverse DTFT)可以定義為：
 
\begin{equation}
x[n] = \frac{1}{2\pi} \int_{-\infty}^{\infty} X(e^{j \omega})e^{j \omega n} d\omega. \tag{17} 
\end{equation}

In general, $X(e^{j \omega})$ is a function of $\omega$ that can be represented as 
 
\begin{equation}
X(e^{j\omega}) = |X(e^ {j\omega})| \cdot e^ {j \theta(\omega)}. \tag{18} 
\end{equation}

where $|X(e^ {j\omega})|$ is magnitude (強度), and $\theta(\omega)$ is argument (幅角) or phase angle (相位角).

Also, if a discrete-time sequence is absolutely summable sequence, then
\begin{equation}
\sum_{n=-\infty}^{\infty} |x[n]| < \infty. \tag{19} 
\end{equation}
Then, we have 
\begin{equation}
|X(e^{j\omega})| < \infty, \text{for all } \omega. \tag{20} 
\end{equation}

#### Example. 
若數位訊號 $x[n]$為：
\begin{equation}
x[n]=\delta [n] 
\end{equation}
or 
\begin{equation}
x[n]=(0.5)^n u[n]
\end{equation}
其中，$\delta [n]$與$u[n]$分別為單位脈衝與單位步階，求離散時間傅立葉轉換(DTFT)。

Sol. (unit pulse)
\begin{equation}
|X(e^{j\omega})| = \sum_{n=-\infty}^{\infty} x[n] \cdot e^{-j\omega n} = \sum_{n=-\infty}^{\infty}\delta[n] \cdot e^{-j\omega n}=1
\end{equation}

Sol. (unit step)

If $\sum_{n=-\infty}^{\infty} |x[n]| < \infty $, then we have 

\begin{equation}
\begin{split}
|X(e^{j\omega})| &= \sum_{n=-\infty}^{\infty} x[n] \cdot e^{-j\omega n} \\
                 &= \sum_{n=-\infty}^{\infty} (0.5)^n u[n] \cdot e^{-j\omega n} \\
                 &= \sum_{n=0}^{\infty} (0.5)^n \cdot e^{-j\omega n} \\
                 &= \sum_{n=0}^{\infty} (0.5 e^{-j\omega})^n=\frac{1}{1-0.5 e^{-j\omega}}
\end{split}
\end{equation}

## <font color='blue'>9-4 DFT (離散傅立葉轉換)</font>
Performing the DFT Using Basic Math<br>
If we have an L-sample-long sequence, $x[n]$, representing the analog continuous-time signal $x(t)$. 
The goal is to find a set of sinusoids which can be added together to produce $x[n]$.

The DFT is based on sampling the DTFT, at equally spaced frequency points. $X(e^{jω})$ is a periodic function in 
$ω$ with a period of 2π. If we take $N$ samples in each period of $X(e^{jω})$, the spacing between frequency points will be 
$\frac{2π}{N}$. 

Hence, the frequency of the set of sinusoids that we are looking for will be of the form $\frac{2π}{N}k$, where we can select 
$k=0,1,...,N−1$. The basis functions will be the form $\frac{2π}{N}k$. 
We are looking for a weighted sum of these functions which will hopefully give us the original signal $x[n]$. 
This means that 
 

\begin{equation}
x[n] = \sum_{k=0}^{N-1} X[k]e^{j \frac{2\pi}{N}kn}, n=0, 1,..., L-1. \tag{21} 
\end{equation}

where $X′(k)$ denotes the weight used for the complex exponential $e^{j \frac{2\pi}{N}kn}$.

Equation 32 is equivalent to the following set of equations 
                                                                      
\begin{equation}
\begin{split}
x[0] &= X'[0] + X'[1] + X'[2] +...+  X'[N-1] \\
x[1] &= X'[0] + X'[1]e^{j \frac{2\pi}{N}} + X'[2]e^{j \frac{2\pi}{N}2} +...+  X'[N-1]e^{j \frac{2\pi}{N}(N-1)} \\
x[2] &= X'[0] + X'[1]e^{j \frac{2\pi}{N}2} + X'[2]e^{j \frac{2\pi}{N}2 \cdot 2} +...+  X'[N-1]e^{j \frac{2\pi}{N}(N-1)\cdot 2}\\
... \\
x[L-1] &= X'[0] + X'[1]e^{j \frac{2\pi}{N}(L-1)} + X'[2]e^{j \frac{2\pi}{N}2(L-1)} +...+  X'[N-1]e^{j \frac{2\pi}{N}(N-1)(L-1)}\\
\end{split}
\end{equation}                                                                 

Given a discrete time series $x[n], n=0, 1,..., N-1$, then the Discrete Fourier Transform (DFT) can be defined as:
 
\begin{equation}
X[k] = \sum_{n=0}^{N-1} x[n]e^{-j \frac{2\pi}{N} kn}, k=0, 1,..., N-1. \tag{22}
\end{equation}

The Inverse DFT can be defined as:
 
\begin{equation}
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{j \frac{2\pi}{N}kn}, n=0, 1,..., N-1. \tag{23}
\end{equation}

It comes from the samples at $\omega$ axis with $\omega= \frac{2\pi}{N} k$

$X[k] = X(j\omega)|_{\omega= \frac{2\pi}{N} k} = \sum_{n=0}^{N-1} x[n] e^{-j\frac{2\pi}{N} kn} $

Therefore, DFT also can be defined as
 
\begin{equation}
X[k] = \sum_{n=0}^{N-1} x[n] W_{N}^{kn}, k=0, 1,..., N-1. \tag{24}
\end{equation}

反離散傅立葉轉換(Inverse DFT)可以定義為：
 
\begin{equation}
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] W_{N}^{kn},n=0, 1,..., N-1. \tag{25}
\end{equation}

where 旋轉因子$ W_{N} = e^{-j (2 \pi/N)}$ and $W_{N}^{kn} = (W_N)^{kn}$

     &= \sum_{m=0}^{\frac{N}{2}-1} x[2m] W_{N}^{k(2m)} + W_N^k\sum_{m=0}^{\frac{N}{2}-1} x[2m+1] W_{N}^{k(2m)}\\

\end{split}
\end{equation}

\begin{equation}

X_e[k+\frac{N}{2}] &= X_e[k] \\
X_o[k+\frac{N}{2}] &= X_o[k] \\
W_N^{k+\frac{N}{2}} &= -W_N^k \\
\end{split}
\end{equation}

That is 
\begin{equation}
\begin{split}
X[k] &= X_e[k] + W_N^k X_o[k]\\
X[k+\frac{N}{2}] &= X_e[k+\frac{N}{2}] + W_N^{k+\frac{N}{2}} X_o[k+\frac{N}{2}] = X_e[k] - W_N^k X_o[k] \\
\end{split}
\end{equation}

X[2] &= \sum_{n=0}^{3} x[n] e^{-j2 \pi (2) n/4} = 1 \cdot e^0 + 2 \cdot e^{-j\pi} + 4\cdot e^{-j2\pi} + 3 \cdot e^{-j3\pi} =0 \\
X[3] &= \sum_{n=0}^{3} x[n] e^{-j2 \pi (3) n/4} = 1 \cdot e^0 + 2 \cdot e^{-j3\pi/2} + 4\cdot e^{-j3\pi} + 3 \cdot e^{-j9\pi/2} \\
     &= -3-j \\
\end{split}
\end{equation} 
Hence, $X=\{10, -3+j, 0, -3-j\}$ <br>
Here, $X[0]$ is a DC componment, while $X[1] - X[3]$ are the AC compoments.
